Now check the percentage of error in each column

```python
null_var = df.isnull().sum()/df.shape[0]*100
null_var
```

```python
drop_column = null_var[null_var > 17].keys()
drop_column
```

```python
df2 = df.drop(columns=drop_column)
df2.shape
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df2.isnull())
```

```python
df3 = df2.dropna(axis=0)
df3.shape
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df3.isnull())
```

```python
df3.isnull().sum().sum()
```
